```python
def convert_height(height):
    feet, inch = map(int, height.replace('"','').split("'"))  # does it convert to integer???? 
    inch = feet*12 + inch
    cm = inch * 2.54
    return cm
    
df1['Height'] = df1['Height'].apply(convert_height)    
df1['Height']
```

```python
def euro_check(monetary_value):
    for char in monetary_value:
        if char not in ('€', 'M', 'K'):
            return True
    return False

df['Value'].apply(euro_check).unique()
```

```python
def covert_value(value):
    if "M" in value:
        new_value = value.replace("€","").replace("M","")
        new_value = float(new_value)*1000000
    elif "K" in value:
        new_value = value.replace("€","").replace("K","")
        new_value = float(new_value)*1000
    else:
        new_value = value.replace("€","")
        new_value = float(new_value)
        
    return new_value

df1['Value'] = df1['Value'].apply(covert_value)
```

```python
def convert_position(nb):
    if isinstance(nb, int):
        return nb 
    
    if '-' in nb:
        score, bonus = map(int, nb.replace('-', '').split("+"))
        total = score - bonus
    else:
        score, bonus = map(int, nb.split("+"))
        total = score + bonus
    
    return total
```

```python
from sklearn.preprocessing import MinMaxScaler 
from sklearn.preprocessing import StandardScaler
```

##### Scaling numerical data, so we can give values to a model

```python
transformer = MinMaxScaler().fit(X_num)
x_standardized = transformer.transform(X_num)

X_num_scaled = pd.DataFrame(x_standardized, columns=X_num.columns)
X_num_scaled.head()
```

##### Encoding categorical values, so they are numrical and we can put them to a model

```python
from sklearn.preprocessing import OneHotEncoder  
```

```python
encoder = OneHotEncoder(drop = "first")  #Creating encoder
encoder.fit(X_cat)  #fitting encoder with objects. 
```

```python
X_cat_encoded = encoder.transform(X_cat).toarray() # Getting values after encoding as arrays
X_cat_encoded
```

```python
X_cat_encoded = pd.DataFrame(X_cat_encoded, columns = encoder.get_feature_names_out()) #Converting to data frame
X_cat_encoded
```

##### Put together all categorical and numerical together again  (Concating features)

```python
df2 = pd.concat([X_cat_encoded, X_num_scaled], axis=1)
```
